```python
import requests

base_url = "http://localhost:8000"
```

### Personal Recommendation

```python
payload = {
    "user_id": "u1",
    "top_k": 3,
    "top_p": 20,
    "content_types":  [
        "series",
        "movie",
        "microdrama",
        "tv",
    ]
}

response = requests.post(f"{base_url}/v1/recommendations", json=payload)
response.json()
```

### Global Recommendation

```python
payload = {
    "top_k": 3,
    "lookback_days": 14,
    "content_types":  [
        "series",
        "movie",
        "microdrama",
        "tv",
    ]
}

response = requests.post(f"{base_url}/v1/popular", json=payload)
response.json()
```
